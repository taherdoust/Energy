```python
# Plot a bar chart to compare 'Comparison_Values_y' and 'elec_consumption' in each month
fig, ax = plt.subplots(figsize=(10, 6))

months = gdf_merged['Month']
comparison_values = gdf_merged['Comparison_Values_y']
elec_consumption_values = gdf_merged['elec_consumption']

bar_width = 0.35
bar_positions = range(len(months))

ax.bar(bar_positions, comparison_values, width=bar_width, label='Comparison_Values_y')
ax.bar([pos + bar_width for pos in bar_positions], elec_consumption_values, width=bar_width, label='elec_consumption')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Values')
plt.title('Comparison of Comparison_Values_y and Elec Consumption for Each Month')
plt.xticks([pos + bar_width/2 for pos in bar_positions], months)
plt.legend()

# Display the plot
plt.show()
```
